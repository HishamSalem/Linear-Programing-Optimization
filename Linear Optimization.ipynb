```python
#Model

import gurobipy as gb
from gurobipy import *
import numpy as np
import pandas as pd


List01=["Ambrosi","Bresciani","Castri","De Blasi","Estensi","Filatoi","Giuliani"]
List02=["Extra_Fine","Fine","Medium","Coarse"]
RHS1 =[2500,3000,2500,2600,2500,38000,2500] 
RHS2=[25000,26000,28000,28000]
Machine_Hours=[[0, 0.400, 0.375, 0.250],
               [0.700,0.500,0.350,0.250],
               [0.675,0.450,0.400,0.250],
               [0,0.450,0.350,0.200],
               [0.650, 0.450,0.400, 0.250],
               [0.625,0.500,0.425,0.425],
               [ 0.700,0.450,0.350,0.400]]

cost_transport = [[0.3, 0.3, 0.45, 0.45],
                  [0.4, 0.4, 0.6, 0.6],
                  [0.8, 0.8, 1.2, 1.2],
                  [0.7, 0.7, 1.05, 1.05],
                  [0.7, 0.7, 1.05, 1.05],
                  [0,0,0,0],
                  [0.5, 0.5, 0.75, 0.75]]

cost_product = [
        [0,13,10.65,9.6],
        [17.4,14.10,11.2,9.45],
        [17.40,14.22,11.00,9.5],
        [0,14.3,11.25,9.6],
        [17.5,13.8,11.4,9.6],
        [18.25,13.9,11.4,8.9],
        [19.75,13.90,10.75,9.4]
        ]

#cost=cost_product+cost_transport
#print(cost)
l = len(List01)
k = len(List02)
I = range(l)
J = range(k)

print(I,J)

model=gb.Model("Filatoi Riuniti")

#Decision Variables

# =============================================================================
# #other way
x = model.addMVar((l,k), lb = 0 , vtype = GRB.CONTINUOUS, name = ["x_"+i+"_"+j for i in List01 for j in List02])
print(x)
cost_product
cost_transport


#objective function

exp = sum(cost_transport[i][j]*x[i][j]+cost_product[i][j]*x[i][j] for i in I for j in J)
model.setObjective(exp, GRB.MINIMIZE)
  
#PRODUCTIONConstraints
model.addConstr(x[0][0]==0,"No Ambrosi Extra Fine yarn production")
model.addConstr(x[3][0]==0,"No De Blasi Extra Fine yarn production")

#for i in I:  
    #model.addConstr(sum(x[i][j] for j in J) <= RHS1[i], "Production"+List01[i])


#Demand Constraints
#for j in J:   
   # model.addConstr(sum(x[i][j] for i in I) >= RHS2[j], "Demand"+List02[j])
    
for i in I:
    model.addConstr(sum(x[i][j] * Machine_Hours[i][j] for j in J) <= RHS1[i], name="Production"+List01[i])

#Adding demand constraint for each type of yarn produced
for j in J:
    model.addConstr(sum(x[i][j] for i in I) >= RHS2[j], name="Demand"+List02[j])

    
#model.getConstrs()
#model.getVars()
model.display()    
model.optimize()

print("Optimal value for Sales/Demand:")
for v in model.getVars():
    print(v.varName, "=", round(v.x,1))

#c
#Print shadow price and constraint slack
o = [{'Constraint':c.ConstrName,
      'Shadow Price':c.Pi,
      'RHS':-c.RHS,
      'Slack': c.Slack}
     for c in model.getConstrs()]
print(pd.DataFrame(o))
    
    
```
